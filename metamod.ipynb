```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Veri setini yükleme
file_path = '/content/drive/MyDrive/melike Yalçın Kaplan/waves.csv'  # Dosya yolunu doğru ayarlayın
df = pd.read_csv(file_path)

# Veri setinin genel yapısı
print("Veri Seti Genel Bilgileri:")
print(df.info())

# İlk 5 satırı inceleme
print("\nVeri Setinin İlk 5 Satırı:")
print(df.head())

# Veri setindeki eksik (NaN) değerlerin sayısı
print("\nEksik Değerlerin Sayısı:")
print(df.isnull().sum())

# Sayısal sütunları seçme
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Sayısal sütunların temel istatistiksel bilgileri
print("\nSayısal Değişkenler için Temel İstatistikler:")
print(df[numeric_columns].describe())

# Korelasyon matrisini sadece sayısal sütunlar üzerinde hesaplama
plt.figure(figsize=(12, 8))
corr_matrix = df[numeric_columns].corr()  # Sayısal sütunlar için korelasyon matrisini oluşturma
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Korelasyon Matrisi")
plt.show()

# Veri setinde yer alan kategorik sütunların dağılımlarını görselleştirme
categorical_columns = df.select_dtypes(include=['object']).columns
plt.figure(figsize=(12, 8))

for i, col in enumerate(categorical_columns, 1):
    plt.subplot(2, 3, i)  # 2 satır, 3 sütunlu bir grid
    sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f'{col} Dağılımı', fontsize=12)

plt.tight_layout()
plt.show()

# Veri setinin sütun türlerinin açıklamaları
print("\nSütun Türleri ve Açıklamaları:")
for col in df.columns:
    print(f"{col}: {df[col].dtype}")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# CSV dosyasını oku
file_path = '/content/drive/MyDrive/melike Yalçın Kaplan/sources.csv'  # Dosyanın yolu
df = pd.read_csv(file_path)

# Sayısal sütunları seçme
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Sayısal sütunlardaki eksik verileri ortalama ile doldurma
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Veri normalizasyonu (sayısal sütunları 0-1 aralığına ölçeklendirir)
scaler = MinMaxScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Temizlenen ve normalleştirilen veri setini kaydetme
output_path = '/content/drive/MyDrive/melike Yalçın Kaplan/cleaned_normalized_sources.csv'
df.to_csv(output_path, index=False)

# Temiz veri setinin ilk birkaç satırını yazdırma
print(df.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Stil ayarı
sns.set(style="whitegrid")

# Örnek veri yükleme (gerçek veriyi buraya yerleştirin)
file_path = '/content/drive/MyDrive/melike Yalçın Kaplan/sources.csv'  # Dosya yolunu doğru ayarlayın
df = pd.read_csv(file_path)

# Sayısal sütunları seçme
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# 2. Sayısal Değişkenler için Dağılım Grafikleri
plt.figure(figsize=(15, 12))

# Grid boyutunu sayısal sütun sayısına göre dinamik olarak ayarlama
n_cols = 5  # Her satırda 5 grafik olacak
n_rows = (len(numeric_columns) // n_cols) + (1 if len(numeric_columns) % n_cols != 0 else 0)

# Alt grafiklerdeki sayılar
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(n_rows, n_cols, i)  # Dinamik olarak satır ve sütun sayısı ayarlandı
    sns.histplot(df[col], kde=True, color='skyblue', bins=30)
    plt.title(col, fontsize=12)

plt.tight_layout()  # Alt satırlara kayma olmasın diye düzenleme
plt.show()
```
